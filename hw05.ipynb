```python
#!/usr/bin/python

import numpy as np

################################### Q1 ###################################
print('#' * 50 + ' Q1 ' + '#'*50)
def power_method(a,max_iteration):
    # choose a random vector
    zt = np.asarray([1,1,1])
    # to save temperal solutions.
    lambda_t_list = [0.0]
    for i in range(max_iteration):
        wt = np.dot(a,zt)
        alpha_t = np.linalg.norm(np.abs(wt))
        k = np.argmax(np.abs(wt))
        lambda_t = wt[k] / zt[k]
        lambda_t_list.append(lambda_t)
        zt = 1./alpha_t * wt # update zt
        error = lambda_t_list[-1] - lambda_t_list[-2]
        print('%02d %.10f %.10e' % (i, lambda_t,error))

a = np.array([[1., 5.,-8.],[5., -2., 5.],[-8., 5.,1.]])
power_method(a, 20)
eigen_vals = np.linalg.eigvals(a)
print('true eigen value lambda_1: %.8f' % eigen_vals[0])
print('true eigen value lambda_2: %.8f' % eigen_vals[1])
print('true eigen value lambda_3: %.8f' % eigen_vals[2])

################################### Q2 ###################################
print('#' * 50 + ' Q2 ' + '#'*50)
a = np.asarray([[ 4, 1, 1, 0, 0],
                [-1,-3, 1, 1, 0],
                [ 2, 1, 5,-1,-1],
                [-1,-1,-1, 4, 0],
                [ 0, 2,-1, 1, 4]])
b = np.asarray( [ 6, 6,-6, 6, 6])
def gauss_seidel_method(a,b,tol=1e-3,max_iter=25):
    x_true = np.linalg.solve(a,b)
    x = np.zeros(5) # initial guess.
    x_next = np.zeros_like(x)
    error_pre = np.inf  
    print('i        x1          x2            x3         x4          x5         error       ratio')
    for i in range(max_iter):
        x_next[0] = 1./4.  * (b[0] - 1.*x[1] - 1.*x[2] + 0.*x[3] + 0.*x[4])
        x_next[1] = 1./-3. * (b[1] + 1.*x_next[0] - 1.*x[2] - 1.*x[3] + 0.*x[4])
        x_next[2] = 1./5.  * (b[2] - 2.*x_next[0] - 1.*x_next[1] + 1.*x[3] + 1.*x[4])
        x_next[3] = 1./4.  * (b[3] + 1.*x_next[0] + 1.*x_next[1] + 1.*x_next[2] + 0.*x[4])
        x_next[4] = 1./4.  * (b[4] - 0.*x_next[0] - 2.*x_next[1] + 1.*x_next[2] - 1.*x_next[3])
        error = np.max(np.abs(x_next-x_true))
        ratio = error / error_pre
        print('%02d %.9f %.9f %.9f %.9f %.9f %.9e %.9f' %(i,x[0],x[1],x[2],x[3],x[4],error,ratio))
        if error < tol:
            break
        x = x_next
        error_pre = error
gauss_seidel_method(a,b)

################################### Q3 ###################################
print('#' * 50 + ' Q3 ' + '#'*50)
a,b,c = 1, -26.0, 1.
x1 = np.sqrt(169.) - np.sqrt(168.)
x2 = np.sqrt(169.) + np.sqrt(168.)
print(x1,x2)
x1 = 1./ (np.sqrt(169.) + np.sqrt(168.))
x2 = np.sqrt(169.) + np.sqrt(168.)
print(x1,x2)
```
